```python

import json
import pandas as pd
import os
```

```python
filename = "countries_extents.json"
countries = []
with open(filename, "r") as file:
    data = json.load(file)
for country_name, extent in data.items():
    if not extent:
        countries.append(country_name)
for country in countries:
    print(country)
```

```python
long_name_to_geo = {
    'Antigua': 'Antigua & Barbuda',
    'North Korea': 'Korea, North',
    'South Korea': 'Korea, South',
    'The Netherlands': 'Netherlands',
    'The Bahamas': 'Bahamas, The',
    'The Gambia': 'Gambia, The',
    'Solomon Islands': 'Solomon Is',
    'Czech Republic': 'Czechia',
    'Bahrain': 'Bosnia & Herzegovina'
}
```

```python
df_extent = pd.read_json("countries.json")
df_extent["long_name"] = df_extent["long_name"].map(lambda x : long_name_to_geo[x] if x in long_name_to_geo else x)
df_geo = pd.read_json("riverCountries.json").rename(columns={0: 'name'})
print(len(df_extent), len(df_geo))
```

```python
df_all = df_geo.merge(df_extent, how='left', left_on=['name'], right_on=['long_name'])
df_all
```

```python
countrie_extent = {}
for idx, row in df_all.iterrows():
    name = row['name']
    if not pd.isna(row['long_name']):
        extent = [row['sw_lng'], row['sw_lat'], row['ne_lng'], row['ne_lat']]
    else:
        extent = []
    countrie_extent[name] = extent
```

```python
with open('countries_extents.json', 'w') as file:
    json.dump(countrie_extent, file)
```
